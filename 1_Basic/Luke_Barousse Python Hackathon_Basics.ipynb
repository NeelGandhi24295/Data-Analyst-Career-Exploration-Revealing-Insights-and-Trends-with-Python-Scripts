## Classes 

```python
class BaseSalary:
    def __init__(self, base_salary, bonus_rate = 0.1, symbol = "$"):
        
        self.base_salary = base_salary  # This self word is for the object itself and self.base_salary is attribute or variables
        self.bonus_rate = bonus_rate
        self.symbol = symbol
        self.total_salary = base_salary * (1 + bonus_rate)
        self.bonus = self.total_salary - self.base_salary
        
# - --- f'{value: , .2f}'' format value with , at tens hundreds, thousands and so on and want 2 decimal places after that



    def __repr__(self):
        return f'{self.symbol}{self.base_salary:,.0f}'

    def show_salary(self):
        return f'{self.symbol}{self.total_salary:,.0f}'

    def show_bonus(self):
        return f'{self.symbol}{self.bonus:,.0f}'
```

```python
salary = BaseSalary(100000)
salary.show_salary()
```

```python
salary.show_bonus()
```
